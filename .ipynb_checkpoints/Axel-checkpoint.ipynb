```python
def standardize_(x):
    """Standardize the original data set."""
    mean_x = np.mean(x, axis=0)
    x = x - mean_x
    std_x = np.std(x, axis=0)
    # Vérifiez si std_x est nul (zéro) avant de diviser
    non_zero_std_indices = std_x != 0
    x[:, non_zero_std_indices] = x[:, non_zero_std_indices] / std_x[non_zero_std_indices]
    return x, non_zero_std_indices

```

Standardization

```python
#Method to split the data features as categorical or numerical

def categorical_and_numerical_data_extraction(dataset, threshold):
    
    categorical_data_index = []
    numerical_data_index = []
    
    for i in range(dataset.shape[1]):
        unique_values = np.unique(dataset[:, i])
        diff_unique_vs_total = (dataset.shape[0]-len(unique_values))/(dataset.shape[0])*100

        if diff_unique_vs_total >= threshold :
            categorical_data_index.append(i)
        
        else:
            numerical_data_index.append(i)
            
    return categorical_data_index, numerical_data_index
        
categorical_data_index, numerical_data_index = categorical_data_extraction(x_train_cleaned, 0.9)  
print(len(categorical_data_index))
print(len(numerical_data_index))

#Deleting the categorical features where there is a huge bias towards one of the category, replacing the nan 
#of the remaining categorical features
def categorical_data_handling(dataset, categorical_data_index, frequency_threshold):
    
    index_to_delete = []
    index_to_modify = []
#dataset[~np.isnan(dataset[:, index])] peut etre a mettre ds le np.unique
    for index in categorical_data_index:
        unique_values, counts = np.unique(dataset[:, index], return_counts = True) #jsp si les nan sont comptés ds le count
        freq_of_each_value = (counts/dataset.shape[0])*100

        if np.any(freq_of_each_value >= frequency_threshold):
            index_to_delete.append(index)
        
        else:
            index_to_modify.append(index)
            
    return index_to_delete, index_to_modify

index_to_delete, index_to_modify = categorical_data_handling(x_train_cleaned, categorical_data_index, 0.95)
print(len(index_to_delete))
print(len(index_to_modify))
```

```python
#Enlever premiere colonne x_train car rempli de nan
x_train = x_train_cleaned[:,0:]
x_test = x_test[:,0:]

#Remplacer les nan par la moyenne de chaque colonne
mean = np.nanmean(x_train, axis = 0)
x_train_cleaned_cleaned = x_train
for col in range(x_train.shape[1]):
    x_train_cleaned_cleaned[:,col] = np.nan_to_num(x_train[:,col], nan= mean[col])

x_train_final, col_to_keep = standardize_(x_train_cleaned_cleaned)

#Enlever les colonnes qui ont une std = 0
x_train_final = x_train_final[:, col_to_keep]
```

Correlation Matrix

```python
correlation_matrix = np.corrcoef(x_train_final, rowvar = False)
plt.figure(figsize=(15, 15))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.show()
```

Deletion of similar columns

```python
#Remplacer la partie triangulaire basse de la matrice par des 0
m = np.tril(np.ones(correlation_matrix.shape), k = 0).astype(bool)
correlation_matrix[m] = 0
```

```python
#Calcul des colonnes qui se ressemblent au dessus d'un threshold
threshold = 0.8

#Creation of a matrix True/False where True == When the correlation is above the treshold
mask = np.logical_and(np.abs(correlation_matrix) >= threshold, np.abs(correlation_matrix) < 1)
correlated_features = np.where(mask == True)

#indices_correlated has the indices of the correlated columns
indices_correlated = np.unique(correlated_features[0])

#Removes the correlated columns to x_train_cleaned
x_train_final = np.delete(x_train_final, indices_correlated, axis = 1)
print(x_train_final.shape)
```
