```python
def standardize(x):
    """Standardize the original data set."""
    mean_x = np.mean(x, axis=0)
    x = x - mean_x
    std_x = np.std(x, axis=0)
    x = x / std_x
    return x, mean_x, std_x
```

```python
#CHeck std =! 0 before
def standardize_(x):
    """Standardize the original data set."""
    mean_x = np.mean(x, axis=0)
    x = x - mean_x
    std_x = np.std(x, axis=0)
    # Vérifiez si std_x est nul (zéro) avant de diviser
    non_zero_std_indices = std_x != 0
    x[:, non_zero_std_indices] = x[:, non_zero_std_indices] / std_x[non_zero_std_indices]
    return x, non_zero_std_indices
```

### Preparation of data : nan -> mean, standardization

```python
#Enlever premiere colonne x_train car rempli de nan
x_train = x_train[:,0:]
x_test = x_test[:,0:]

#x_train_cleaned = np.nan_to_num(x_train, nan=0.0)
mean = np.nanmean(x_train, axis = 0)
x_train_cleaned = x_train
for col in range(x_train.shape[1]):
    x_train_cleaned[:,col] = np.nan_to_num(x_train[:,col], nan= mean[col])

x_train_final, col_to_keep = standardize_(x_train_cleaned)
x_train_final = x_train_final[:, col_to_keep]
print(x_train_final.shape)
```
