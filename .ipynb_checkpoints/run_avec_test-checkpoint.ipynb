```python
#check if duplicates in x_train
duplicata(x_train)    

#shuffle x_train and y_train at the same time to shuffle them in the same way
x_train1, y_train1 = shuffle_rows (x_train, y_train)

#check the distribution of sick and not sick
test_balancy_dataset(y_train1)
    
#premiere colonne remplie de Nan (supprimer d autres des premieres colonnes)
x_train1 = x_train1[:,9:]
x_test1 = x_test[:,9:]

#remplacer Nan par mediane (estce qu on a le droit de faire ca dans xtest? estce que cest une bonne idee ?)
x_train1=nan_to_median(x_train1)
x_test1=nan_to_median(x_test1)

#enlever les lignes d outliers pour chaque colonne
#x_train1, y_train1 = remove_outliers(x_train1, y_train1, threshold=0.000002)

#standardization of test
x_test1, _ = standardize_(x_test1)

#enlever les colonnes qui ont une std nulle
x_train1, col_to_keep=remove_zero_std_columns(x_train1)
x_test1= x_test1[:, col_to_keep]
print('Shape x_train after standardization : ', x_train1.shape)
print('Shape y_train after standardization: ', y_train1.shape)

#enlever les colonnes avec trop grande correlation
x_train1, col_to_delete=remove_high_correlation_feature(x_train1)
x_test1= np.delete(x_test1, col_to_delete, axis = 1)
print('Shape x_train after corr : ', x_train1.shape)
print('Shape y_train after corr: ', y_train1.shape)

#recuperer les indices des colomnes numeriques et categoriques (au dessus de 5 valeurs diff on considere numerical)
#categorical_columns, numeric_columns, _, _=separate_column_types(x_train1, max_nb=5)

#dans les colonnes numeriques, supprimer les lignes correspondant  des outliers
#x_train1, y_train1=remove_numerical_outliers(x_train1, y_train, numeric_columns, z_threshold=15)
#print('Shape x_train after outlier removal : ', x_train1.shape)
#print('Shape y_train after outlier removal : ', y_train1.shape)

#enlever les lignes de outliers
#x_train1, rows_to_keep =remove_outlier_rows(x_train1)

#remplacer les outliers dans les colonnes qu'il reste avec la medianne aussi (le faire pour x_test?)
#x_train1=replace_remaining_outlier_values(x_train1)
#x_test1 = replace_remaining_outlier_values(x_test1)

#Add a column of ones TO USE ONLY WITH SUBSETS
x_train1 = build_model_data(x_train1)
x_test1 = build_model_data(x_test1)

print('Finale shape of x_train : ', x_train1.shape)
print('Finale shape of y_train : ', y_train1.shape)
print('Finale shape of x_test : ', x_test1.shape)
```

# 4- First step w

### a) Least Squares

```python
w_initial, _ = least_squares(y_tr, x_tr)  #BEST THRESHOLD = 0.36

test(w_initial, x_te, y_te, threshold = 0.36)
```

With subsets:

```python
weights = []

for i in range(len(sub_x_trains)):
    w,_ = least_squares(sub_y_trains[i], sub_x_trains[i])
    weights.append(w)
    
w_initial = np.mean(weights, axis = 0)

test(w_initial, x_te, y_te, threshold = 0.59)  #BEST THRESHOLD = 0.59
```

### b) Ridge Regression

```python
lambda_ = 0.001
w_initial, loss = ridge_regression(y_tr, x_tr, lambda_)   

test(w_initial, x_te, y_te, threshold = 0.36)  #BEST THRESHOLD = 0.36
```

With subsets and fixed lambda :

```python
weights = []
lambda_ = 0.00005

for i in range(len(sub_x_trains)):
    w,_ = ridge_regression(sub_y_trains[i], sub_x_trains[i], lambda_)
    weights.append(w)
    
w_initial = np.mean(weights, axis = 0)

test(w_initial, x_te, y_te, threshold = 0.6)  #BEST THRESHOLD = 0.6
```

##### Search for good hyperparameters  lambda_ for ridge regression

```python
lambdas = np.linspace(0,0.003,20)
weights = []
losses_te = []
f = []
acc = []

for lambda_ in lambdas:
    w,_ = ridge_regression(y_tr, x_tr, lambda_)
    weights.append(w)
    loss = compute_loss(y_te, x_te, w)
    losses_te.append(loss)
    f.append(compute_f1(y_te, sigmoid_prediction_ (x_te, w, threshold=0.36)))
    acc.append(accuracy(y_te, sigmoid_prediction_ (x_te, w, threshold=0.36)))

best_index = losses_te.index(min(losses_te))
print('Smallest loss : ',losses_te[best_index])
print('Best lambda : ',lambdas[best_index])
w_initial = weights[best_index]

test(w_initial, x_te, y_te, losses = losses_te, f1_scores = f, accuracies = acc, gammas = None, lambdas = lambdas, threshold = 0.36)
```

##### Search of hyperparameter lambda with subsets

```python
lambdas = np.linspace(0,0.001,20)   #BEST LAMBDA = 0.00005
weights = []
losses_te = []
acc = []
f = []

for lambda_ in lambdas:
    weights_lambda = []
    for i in range(len(sub_x_trains)):
        w,_ = ridge_regression(sub_y_trains[i], sub_x_trains[i], lambda_)
        weights_lambda.append(w)
        
    weights.append(np.mean(weights_lambda, axis = 0))
    loss = compute_loss(y_te, x_te, np.mean(weights_lambda, axis = 0))
    losses_te.append(loss)
    f.append(compute_f1(y_te, sigmoid_prediction_ (x_te, np.mean(weights_lambda, axis = 0), threshold=0.6)))
    acc.append(accuracy(y_te, sigmoid_prediction_ (x_te, np.mean(weights_lambda, axis = 0), threshold=0.6)))

best_index = losses_te.index(min(losses_te))
print('Smallest loss : ',losses_te[best_index])
print('Best lambda : ',lambdas[best_index])
w_initial = weights[best_index]

test(w_initial, x_te, y_te, losses = losses_te, f1_scores = f, accuracies = acc, gammas = None, lambdas = lambdas, threshold = 0.6)
```

# 5 - Gradient descent

### a) Mean squared error gradient descent

```python
#With fixed hyperparamaters
max_iters = 10
gamma = 0.1

w_final, _ = mean_squared_error_gd(y_tr, x_tr, w_initial, max_iters, gamma)

test(w_final, x_te, y_te, threshold = 0.407)  #BEST THRESHOLD = 0.407
```

##### With subsets

```python
weights = []
gamma = 0.01
max_iters = 30

w = w_initial
for i in range(len(sub_x_trains)):
    w,_ = mean_squared_error_gd(sub_y_trains[i], sub_x_trains[i], w, max_iters, gamma)
    weights.append(w)
    
w_final = np.mean(weights, axis = 0)

test(w_final, x_te, y_te, threshold = 0.6) #BEST THRESHOLD = 0.6
```

##### Search of best hyperparameter gamma

```python
losses_te = []
weights = []
f = []
acc = []
max_iters = 10
gammas = np.linspace(0,1,50)

w = w_initial

for gamma in gammas :
    w, loss = mean_squared_error_gd(y_tr, x_tr, w, max_iters, gamma)
    weights.append(w_final)
    
    losses_te.append(compute_loss(y_te, x_te, w_final))
    f.append(compute_f1(y_te, sigmoid_prediction_ (x_te, w, threshold=0.6)))
    acc.append(accuracy(y_te, sigmoid_prediction_ (x_te, w, threshold=0.6)))

best_index = losses_te.index(min(losses_te))
print('Best gamma : ', gammas[best_index])
print('Smallest loss : ', losses_te[best_index])
w_final = weights[best_index]   

test(w_final, x_te, y_te, losses = losses_te, f1_scores = f, accuracies = acc, gammas = gammas, lambdas = None, threshold = 0.6)
```

##### Search best hyperparameters with subsets

```python
losses_te = []
weights = []
f = []
acc = []
max_iters = 10
gammas = np.linspace(0,1,50)

for gamma in gammas :
    weights_sub = []
    for i in range(len(sub_x_trains)):
        w = w_initial
        w, loss = mean_squared_error_gd(sub_y_trains[i], sub_x_trains[i], w, max_iters, gamma)
        weights_sub.append(w)
    weights.append(np.mean(weights_sub, axis = 0))
    losses_te.append(compute_loss(y_te, x_te, np.mean(weights_sub, axis = 0)))
    f.append(compute_f1(y_te, sigmoid_prediction_ (x_te, np.mean(weights_sub, axis = 0), threshold=0.6)))
    acc.append(accuracy(y_te, sigmoid_prediction_ (x_te, np.mean(weights_sub, axis = 0), threshold=0.6)))

best_index = losses_te.index(min(losses_te))
print('Best gamma : ', gammas[best_index])
print('Smallest loss : ', losses_te[best_index])
w_final = weights[best_index]   

test(w_final, x_te, y_te, losses = losses_te, f1_scores = f, accuracies = acc, gammas = gammas, lambdas = None, threshold = 0.6)
```

##### Search of best hyperprameters lambda and gamma for ridge regression AND regularized logistic regression without subsets

```python
nb_steps = 20
lambdas = np.linspace(0,0.4,10)
gammas = np.linspace(0,1,10)
weights = np.zeros((len(lambdas),len(gammas), x_tr.shape[1]))
losses_te = np.zeros((len(lambdas),len(gammas)))

for i, lambda_ in enumerate(lambdas):
    for j, gamma in enumerate(gammas):
        
        w, _ = ridge_regression(y_tr, x_tr, lambda_)

        for step in range(nb_steps):

            w, loss = reg_logistic_regression(y_tr, x_tr, w, gamma, lambda_)

        weights[i,j] = w
        loss_te = compute_loss(y_te, x_te, w)
        losses_te[i][j] = loss_te   

best_index = np.unravel_index(np.argmin(losses_te), losses_te.shape)
row_index, col_index = best_index
best_weight = weights[row_index][col_index]
print('Best lambda : ', lambdas[row_index])
print('Best gamma : ', gammas[col_index])
print('Smallest loss : ', losses_te[best_index])
w_final = best_weight

test(w_final, x_te, y_te, threshold = 0.5)
```

Search of best hyperparameters with subsets

```python
nb_steps = 20
lambdas = np.linspace(0,0.4,10)
gammas = np.linspace(0,1,10)
weights = np.zeros((len(lambdas),len(gammas), x_tr.shape[1]))
losses_te = np.zeros((len(lambdas),len(gammas)))

for i, lambda_ in enumerate(lambdas):
    for j, gamma in enumerate(gammas):
        w = w_initial
        weights_sub = []
        for a in range(len(sub_x_trains)):
            
            for step in range(nb_steps):

                w, _ = reg_logistic_regression(sub_y_trains[a], sub_x_trains[a], w, gamma, lambda_)
            weights_sub.append(w)

        weights[i,j] = np.mean(weights_sub, axis = 0)
        loss_te = compute_loss(y_te, x_te, np.mean(weights_sub, axis = 0))
        losses_te[i][j] = loss_te   

best_index = np.unravel_index(np.argmin(losses_te), losses_te.shape)
row_index, col_index = best_index
best_weight = weights[row_index][col_index]
print('Best lambda : ', lambdas[row_index])
print('Best gamma : ', gammas[col_index])
print('Smallest loss : ', losses_te[best_index])
w_final = best_weight

test(w_final, x_te, y_te, threshold = 0.6)
```
